```python
import numpy as np
import pandas as pd

import matplotlib # Para ver la versión
import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter

import sklearn # Para ver la versión

from sklearn import datasets

# Escalado
from sklearn.preprocessing import MinMaxScaler
```

```python
df_final_clustering = pd.read_csv(r"df_final_clustering.csv")
```

```python
df_final_clustering
```

```python
df_final_clustering_2 = df_final_clustering.copy()

df_final_clustering_groupby = df_final_clustering_2.groupby(by = "estudios", as_index = False).agg({"salario_medio" : "median"}).sort_values("salario_medio").dropna()

plt.figure(figsize = (12, 4))
sns.boxplot(x = df_final_clustering_2.estudios, y = df_final_clustering_2.salario_medio, order = df_final_clustering_groupby.estudios)
plt.xticks(rotation = 90)
plt.show()
```
