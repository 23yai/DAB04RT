```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
from datetime import datetime
```

```python
reps = int(BeautifulSoup(requests.get('https://www.tecnoempleo.com/ofertas-trabajo/?pagina=1').text, "html.parser").find("h1").text.split()[0].replace(".",""))
links = [f"https://www.tecnoempleo.com/ofertas-trabajo/?pagina={i+1}" for i in range(reps//30+1)]
links
```

```python
urls = []
oferta = []
empresa = []

for i in links:
    response = requests.get(i)
    soup = BeautifulSoup(response.text, "html.parser")
    
    for l in soup.find_all("div", class_ = "row fs--15"):
        try:
            oferta.append(l.find("h3").find("a")["title"])
        except:
            oferta.append(np.nan)


    for k in soup.find_all("div", class_ = "row fs--15"):
        try:
            empresa.append(k.find("a", class_ = "text-primary link-muted").text)
        except:
            empresa.append(np.nan)
        
    for j in soup.find_all("h3"):
        urls.append(j.find("a")["href"])
        
    sleep(2)        
    print(i)
```

```python
df = pd.DataFrame()
df["Urls"] = urls
df["Oferta"] = oferta
df["Empresa"] = empresa
```

```python
df
```

```python
fecha_de_publicacion = []
aptitudes = []
funcion = []
requisito = []
salario = []
contrato = []
jornada = []
idiomas = []
ubicacion = []
texto = []
experiencia = []
fecha_extraccion = []

for i in df["Urls"]:
    response = requests.get(i)
    soup = BeautifulSoup(response.text, "html.parser")
    numeros = []
    formacion = False
    contra = False
    jorn = False
    idiom = False
    puesto = False
    ubi = False
    text = False
    exp = False


    # EXTRACCION DE DATOS
    
    fecha_extraccion.append(datetime.now())
    
    try:
        fecha_de_publicacion.append(soup.find("span", class_ = "ml-4").text.strip()[:10])
    except:
        fecha_de_publicacion.append(np.nan)

    
    try:
        aptitudes.append(soup.find("div", class_= "pl--12 pr--12").text.split())
    except:
        aptitudes.append(np.nan)
        

    for k in soup.find_all("div", class_ = "mt-4"):
        if k.find("p", class_ = "fs--16 text-gray-800"):
            texto.append(" ".join(k.text.split()).strip())
            text = True
            break
    if not text:
        texto.append(np.nan)

    
    for k in soup.find_all("li", class_ = "list-item clearfix border-bottom py-2"):
        if k.find("i", class_ = "fi fi-users"):
            funcion.append(" ".join([" ".join(i.replace("Funciones", "").split()) for i in (k.text.split())]).strip())
            puesto = True
            break
    if not puesto:
        funcion.append(np.nan)
        
    
    for k in soup.find_all("div", class_ = "d-flex py-2"):
        if k.find("i", class_ = "fi fi-graduate"):
            requisito.append((k.text.split(":")[1]).strip())
            formacion = True
            break
    if not formacion:
        requisito.append(np.nan)

    
    for l in soup.find_all("li", class_ = "list-item clearfix border-bottom py-2"):
        if l.find("i", class_="fi fi-plus"):
            for k in l.text.strip().replace(".","")[:-1].split():
                if k.isdigit():
                    numeros.append(k)
    if numeros:
        salario.append(" ".join(numeros))
    else: 
        salario.append(np.nan)

    
    for l in soup.find_all("li", class_ = "list-item clearfix border-bottom py-2"):
        if l.find("i", class_ = "fi fi-task-list"):
            contrato.append(" ".join([" ".join(i.replace("Tipo", "").replace("contrato", "").split()) for i in (l.text.split())]).strip())
            contra = True
            break
    if not contra:
        contrato.append(np.nan)

    
    for l in soup.find_all("li", class_ = "list-item clearfix border-bottom py-2"):
        if l.find("i", class_ = "fi fi-calendar"):
            jornada.append(" ".join(l.text.split()[:2]).strip())
            jorn = True
            break
    if not jorn:
        jornada.append(np.nan)
        

    for l in soup.find_all("div", class_ = "d-flex py-2"):
        if l.find("i", class_ = "fi fi-globe"):
            idiomas.append(" ".join((l.text.replace("|", "").split(":")[1]).split()))
            idiom = True
            break
    if not idiom:
        idiomas.append(np.nan)


    for k in soup.find_all("li", class_ = "list-item clearfix border-bottom py-2"):
        if k.find("i", class_ = "fi fi-pin"):
            ubicacion.append(" ".join(k.text.split()).strip())
            ubi = True
            break
    if not ubi:
        ubicacion.append(np.nan)
        

    for k in soup.find_all("li", class_ = "list-item clearfix border-bottom py-2"):
        if k.find("i", class_ = "fi fi-shield-ok"):
            experiencia.append(" ".join(k.text.split()).replace("Experiencia", "").strip())
            exp = True
            break
    if not exp:
        experiencia.append(np.nan)
        
    sleep(1)

print(len(urls),
      len(oferta),
      len(empresa),
      len(fecha_de_publicacion),
      len(aptitudes),
      len(funcion),
      len(requisito),
      len(salario),
      len(contrato),
      len(jornada),
      len(idiomas),
      len(ubicacion),
      len(texto),
      len(experiencia),
      len(fecha_extraccion))
```

```python
## De esta manera crearemos la columna de Teletrabajo o Presencial con la de Ubicacion.
## Al establecer una funcion podremos hacer uso de ella mas adelante para los nuevos urls
def ext_teletrabajo(df_ubicacion):
    teletrabajo = []
    for i in df_ubicacion.values:
        if not pd.isna(i):
            if "Presencial" in i:
                teletrabajo.append("Presencial")
                
            elif "remoto" in i:
                teletrabajo.append("Remoto")
                
            elif "(Híbrido)" in i:
                teletrabajo.append("Híbrido")
                
            else:
                teletrabajo.append(np.nan)
        else:
            teletrabajo.append(i)
        
    return teletrabajo
```

```python
df["Funcion"] = funcion
df["Aptitudes"] = aptitudes
df["Requisito"] = requisito
df["Idiomas"] = idiomas
df["Jornada"] = jornada
df["Ubicacion"] = ubicacion
```

```python
df["Teletrabajo o Presencial"] = ext_teletrabajo(df["Ubicacion"])
df
```

```python
df["Contrato"] = contrato
df["Salario"] = salario
df["Experiencia"] = experiencia
df["Texto"] = texto
df["Fecha_de_publicacion"] = fecha_de_publicacion
df["Fecha_extraccion"] = fecha_extraccion
```

```python
## Una vez creado el df teletrabajo podemos limpiar esa informacion del df ubicacion
df["Ubicacion"] = df["Ubicacion"].map(lambda x: x.replace("(Híbrido)","").replace("100% En remoto", "Teletrabajo").replace("(Presencial)","").strip() if not pd.isna(x) else x)
df["Ubicacion"]
```
