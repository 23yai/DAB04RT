```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
```

```python
ofertas_empleo = pd.read_csv("C:\\Users\\GoraJ\\Downloads\\aprendiendo python\\PFB\\limpieza\\df_final.csv",
                      dtype={"salario_min": "float64",
                             "salario_max": "float64",
                             "estudios": "bool",
                             "experiencia": "float64",
                             "vacaciones": "Int64"},
                      parse_dates=["fecha", "fecha_extraccion"])
```

```python
ofertas_empleo = ofertas_empleo.drop(["id_urls", "fecha", "oferta", "fecha_extraccion", "descripcion", "empresa"], axis = 1)
```

```python
ofertas_empleo
```

```python
#aqui se pone el numero de skills/tecnologias_aptitudes/beneficios que piden porque son varias
ofertas_empleo["skills"] = ofertas_empleo["skills"].apply(lambda x : len(x.split(",")) if not pd.isna(x) else 0)
ofertas_empleo["tecnologias_aptitudes"] = ofertas_empleo["tecnologias_aptitudes"].apply(lambda x : len(x.split(",")) if not pd.isna(x) else 0)
ofertas_empleo["beneficios"] = ofertas_empleo["beneficios"].apply(lambda x : len(x.split(",")) if not pd.isna(x) else 0)
```

```python
ofertas_empleo
```

```python
#VARIABLES YA NUMERICAS (SALARIO, EXPERIENCIA, ESTUDIOS)
# salario medio
ofertas_empleo["salario_medio"] = ofertas_empleo[["salario_min", "salario_max"]].mean(axis=1)

# experiencia 
ofertas_empleo["experiencia"] = ofertas_empleo["experiencia"].fillna(0)

# bool a 0 y 1
ofertas_empleo["estudios"] = ofertas_empleo["estudios"].astype(int)
```

```python
ofertas_empleo
```

```python
#aqui se ponen por lista las ubicaciones e idiomas
ofertas_empleo["ubicacion"] = ofertas_empleo["ubicacion"].apply(lambda x : [y.strip() for y in x.split(",")] if not pd.isna(x) else x)
ofertas_empleo = ofertas_empleo.explode("ubicacion").reset_index(drop = True)

ofertas_empleo["idiomas"] = ofertas_empleo["idiomas"].apply(lambda x : [y.strip() for y in x.split(",")] if not pd.isna(x) else x)
ofertas_empleo = ofertas_empleo.explode("idiomas").reset_index(drop = True)
```

```python
ofertas_empleo
```

```python
#antes de clasificar españa/extranjero voy a rellenar los nans de modalidad en funcion de la ubicacion
#si la ubicacion es teletrabajo o nan, en modalidad se pone "remoto" y si no es ninguna de estas es "presencial"
ofertas_empleo.loc[(ofertas_empleo["modalidad"].isna()) & 
                    ((ofertas_empleo["ubicacion"] == "Teletrabajo") | (ofertas_empleo["ubicacion"].isna())), "modalidad"] = "Remoto"

ofertas_empleo.loc[(ofertas_empleo["modalidad"].isna()) & 
                    ~((ofertas_empleo["ubicacion"] == "Teletrabajo") | (ofertas_empleo["ubicacion"].isna())), "modalidad"] = "Presencial"
```

```python
ofertas_empleo["ubicacion"].unique()
```

```python
#aqui se utilizan dos opciones (ubicacion: españa/extrajnero)
ubicaciones_extranjero = ['Malta', 'Holanda', 'Portugal', 'Reino Unido', 'Dinamarca', 'Polonia', 
                          'Alemania', 'Bélgica', 'Suecia', 'Andorra', 'Francia', 'Israel',
                          'Luxemburgo', 'Rumanía', 'Italia', 'Macedonia', 'Bulgaria', 'Irlanda',
                          'Hungría', 'London', 'Chipre', 'Grecia', 'Norte Europa', 'Utrecht']

# se reemplaza la columna por la clasificación españa/extranjero
ofertas_empleo["ubicacion"] = ofertas_empleo["ubicacion"].apply(lambda x: "Extranjero" if x in ubicaciones_extranjero else "España")
```

```python
ofertas_empleo["idiomas"].unique()
```

```python
#aqui se utilizan dos opciones (idiomas: ingles/otros)
ofertas_empleo["idiomas"] = ofertas_empleo["idiomas"].apply(lambda x: "Inglés" if x == "Inglés" else "Otros")
```

```python
ofertas_empleo
```

```python
#se mapean las funciones del puesto en funcion del salario_medio
map_funcion = {k : v for k,v in ofertas_empleo.groupby(by = "funcion", as_index = False).agg({"salario_medio" : "mean"}).values}
ofertas_empleo.funcion = ofertas_empleo.funcion.map(map_funcion)
```

```python
ofertas_empleo
```

```python
ofertas_empleo = ofertas_empleo.drop(["salario_min", "salario_max"], axis = 1)
```

```python
ofertas_empleo
```

```python
#VARIABLES CATEGORICAS (MODALIDAD, JORNADA, CONTRATO)

# modalidad teletrabajo/remoto/híbrido
ofertas_empleo["modalidad"] = ofertas_empleo["modalidad"].replace({"Teletrabajo" : "Remoto"})
#ESTO YA NO DEBERÍA PORQUE SE #ESTO YA NO DEBERÍA PORQUE SE VA A REEMPLAZADO ANTERIORMENTE EN FUNCION DE LA UBICACION
#ofertas_empleo["modalidad"] = ofertas_empleo["modalidad"].fillna("Desconocido")

#jornada, rellenamos nans con desconocida ESTO YA NO DEBERÍA PORQUE SE VA A RELLENAR CON KNN
#ofertas_empleo["jornada"] = ofertas_empleo["jornada"].fillna("Desconocida")

# reemplzar los que no sean obvios por "otros"
ofertas_empleo["contrato"] = ofertas_empleo["contrato"].replace({"Freelance/Autónomo" : "Otros",
                                                                 "Obra o servicio" : "Otros",
                                                                 "A determinar" : "Otros"})
#ESTO YA NO DEBERÍA PORQUE SE VA A RELLENAR CON KNN
#ofertas_empleo["contrato"] = ofertas_empleo["contrato"].fillna("Otros")
```

```python
ofertas_empleo
```

```python
#map de PORTAL EMPLEO
ofertas_empleo["portal_empleo"] = ofertas_empleo["portal_empleo"].map({"tecnoempleo": 0,
                                                                       "getmanfred": 1})
```

```python
ofertas_empleo
```

voy a ir haciendo **KNN** de las columnas numericas funcion/vacaciones/salario_medio

```python
#columnas relacionadas con FUNCION
columnas_para_knn = ["funcion", "tecnologias_aptitudes", "experiencia", "estudios"]

#aislar solo esas columnas
df_knn = ofertas_empleo[columnas_para_knn]

#knn imputer
imputer = KNNImputer(n_neighbors=5)

#imputar los datos para ajustarlos
df_knn_imputado = imputer.fit_transform(df_knn)

#reemplazar los datos de esa columna
ofertas_empleo["funcion"] = df_knn_imputado[:, 0]
```

```python
#columnas relacionadas con VACACIONES
columnas_para_knn = ["vacaciones", "funcion", "experiencia"]

#aislar solo esas columnas
df_knn = ofertas_empleo[columnas_para_knn]

#knn imputer
imputer = KNNImputer(n_neighbors=5)

#imputar los datos para ajustarlos
df_knn_imputado = imputer.fit_transform(df_knn)

#reemplazar los datos de esa columna
ofertas_empleo["vacaciones"] = df_knn_imputado[:, 0]
```

```python
#columnas relacionadas con SALARIO_MEDIO
columnas_para_knn = ["salario_medio", "experiencia", "estudios", "tecnologias_aptitudes"]

#aislar solo esas columnas
df_knn = ofertas_empleo[columnas_para_knn]

#knn imputer
imputer = KNNImputer(n_neighbors=5)

#imputar los datos para ajustarlos
df_knn_imputado = imputer.fit_transform(df_knn)

#reemplazar los datos de esa columna
ofertas_empleo["salario_medio"] = df_knn_imputado[:, 0]
```

```python
ofertas_empleo
```

en principio faltaria **quitar los nans de jornada y contrato**, al tener poco vol. de nans y no afectar significativamente en los resultados, vamos a rellenarlo con la **moda**

```python
# moda jornada
moda_jornada = ofertas_empleo["jornada"].mode()[0]
ofertas_empleo["jornada"] = ofertas_empleo["jornada"].fillna(moda_jornada)

# moda contrato
moda_contrato = ofertas_empleo["contrato"].mode()[0]
ofertas_empleo["contrato"] = ofertas_empleo["contrato"].fillna(moda_contrato)
```

```python
ofertas_empleo
```

```python
#columnas con mapeo 1/0: IDIOMAS/UBICACION
ofertas_empleo["idiomas"] = ofertas_empleo["idiomas"].map({"Inglés": 0,
                                                           "Otros": 1})
ofertas_empleo["ubicacion"] = ofertas_empleo["ubicacion"].map({"España": 0,
                                                               "Extranjero": 1})
```

```python
ofertas_empleo
```

```python
#ahora sí el ultimo paso sería hacer el one-hot encoding de las columnas categóricas que quedan MODALIDAD/JORNADA/CONTRATO
col_categoricas = ["modalidad", "jornada", "contrato"]

#iniciamos el encoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown="error")

#aplicamos en las col categoricas
encoded_values = encoder.fit_transform(ofertas_empleo[col_categoricas])

#las ponemos en un dataframe
encoded_df = pd.DataFrame(encoded_values, columns=encoder.get_feature_names_out(col_categoricas))

#las concatenamos a nuestro dataframe
df_final_clustering = pd.concat([ofertas_empleo.drop(columns=col_categoricas), encoded_df], axis=1)
```
