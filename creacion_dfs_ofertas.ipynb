```python
df_tecnologias = pd.DataFrame(set(DF_MILA["Tecnolog√≠as"].dropna().apply(eval).sum())).reset_index()

df_tecnologias.columns = ["id_tecnologia", "tecnologia"]

df_tecnologias["id_tecnologia"] = df_tecnologias["id_tecnologia"] + 1

df_tecnologias
```

```python
df_oferta_tecnologia = DF_MILA[["url", "Tecnolog√≠as"]].dropna()

df_oferta_tecnologia["Tecnolog√≠as"] = df_oferta_tecnologia["Tecnolog√≠as"].apply(eval)

df_oferta_tecnologia = df_oferta_tecnologia.explode("Tecnolog√≠as").reset_index(drop = True)

map_tecnologias = {v : k for k,v in df_tecnologias.values}

df_oferta_tecnologia["Tecnolog√≠as"] = df_oferta_tecnologia["Tecnolog√≠as"].map(map_tecnologias)

df_oferta_tecnologia.columns = ["id_url", "id_tecnologia"]

df_oferta_tecnologia = df_oferta_tecnologia.dropna().reset_index(drop = True)

df_oferta_tecnologia["id_tecnologia"] = df_oferta_tecnologia["id_tecnologia"].apply(int)

df_oferta_tecnologia
```
