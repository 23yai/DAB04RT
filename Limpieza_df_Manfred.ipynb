```python
meses = {
    "ene": "01", "feb": "02", "mar": "03", "abr": "04", "may": "05", "jun": "06",
    "jul": "07", "ago": "08", "sep": "09", "oct": "10", "nov": "11", "dic": "12"
}
```

```python
#funcion para convertir la fecha en formato MM/YYYY
def convertir_fecha(fecha):
    if isinstance(fecha, str) and len(fecha) == 5:
        mes_abr = fecha[:3].lower()
        anio = "20" + fecha[3:]
        mes = meses.get(mes_abr)
        if mes:
            return f"{mes}/{anio}"
    return None
```

```python
#aplicar las funciones a las columnas
df2["fecha"] = df2["fecha"].apply(convertir_fecha)
```

```python
import pandas as pd

# Suponiendo que 'df' es tu DataFrame y 'salario' es la columna que contiene los salarios.
def limpiar_salario(salario):
    # Eliminar los s√≠mbolos de moneda y cualquier espacio extra
    salario = str(salario).replace("‚Ç¨", "").replace("US$", "").replace("¬£", "").replace(",", "").replace("K", "").replace("Hasta", "").replace("Desde", "").replace("/hora", "").strip()

    # Verificar si el salario contiene un rango (con guion)
    if '-' in salario:
        salario_min, salario_max = salario.split('-')
        # Convertir a valores num√©ricos
        return pd.Series([float(salario_min)*1000, float(salario_max)*1000])
    else:
        # Si no hay rango, asignar el mismo valor a ambas columnas
        return pd.Series([float(salario)*1000, float(salario)*1000])

# Aplicar la funci√≥n para limpiar los salarios
df2[['salario_min', 'salario_max']] = df2['salario'].apply(limpiar_salario)

# Mostrar el DataFrame resultante
df2.head()
```
